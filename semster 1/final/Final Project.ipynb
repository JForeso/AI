```python
import pandas as pd
import seaborn as sns
import numpy as np
import requests
from bs4 import BeautifulSoup
import random
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame()
count=0
page = 1
while count != 30:
    count+=1
    page+=1
    url= 'https://api.rawg.io/api/games?key=026c1e70881c4308a3c3d8ca1a870c5d&page_size=40&page='+ str(page)
    r = requests.get(url)
    data = r.json()
    df = df.append(pd.DataFrame(data["results"]), ignore_index=True)
```

```python
df
```

```python
df.info()
```

```python

df["genres"] = df["genres"].astype(str)
```

```python
dfa=df[df['genres'].str.contains('action', case=False)]
```

```python
dfs=df[df['genres'].str.contains('Strategy', case=False)]
```

```python
dfa
```

```python
dfa['source'] = 'DF1'  
dfs['source'] = 'DF2'
```

```python
combined_df = pd.concat([dfa, dfs])
```

```python
sns.barplot(x='gen', y='rating', data=combined_df, hue='source', palette=['blue', 'orange'])
```

```python
dfs['rating'].mean()
```

```python
dfa['rating'].mean()
```

```python
dfa5=dfa[dfa['rating_top'] == 5].shape[0]
dfaa=dfa.shape[0]
```

```python
dfs5=dfs[dfs['rating_top'] == 5].shape[0]
dfaas=dfs.shape[0]
```

```python
(dfa5/dfaa)*100
```
