```python
import pandas as pd
import seaborn as sns
import numpy as np
import requests
from bs4 import BeautifulSoup
import random
```

```python
df = pd.DataFrame()
count=0
page = 1
while count != 30:
    count+=1
    page+=1
    url= 'https://api.rawg.io/api/games?key=026c1e70881c4308a3c3d8ca1a870c5d&page_size=40&page='+ str(page)
    r = requests.get(url)
    data = r.json()
    df = df.append(pd.DataFrame(data["results"]), ignore_index=True)
```

```python
df
```

```python
df.info()
```

```python
df["released"]= pd.to_datetime(df['released'])
```

```python
sns.regplot(data=df, x=df['rating'], y=df['ratings_count'],color='red' )
```

```python
correlation_matrix = df.corr(method='pearson')

print(correlation_matrix)
```
